```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
x=df.drop(['Price'],axis=1)
y=df['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x_train.info()
```

```python
x_train.columns
```

```python
x_train['Weight'].unique()
```

```python
x_train['Weight']=x_train['Weight'].str.replace('kg','').astype(float)
x_test['Weight']=x_test['Weight'].str.replace('kg','').astype(float)
x_train['Ram']=x_train['Ram'].str.replace('GB','').astype(int)
x_test['Ram']=x_test['Ram'].str.replace('GB','').astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle
le=LabelEncoder()

```

```python
for col in x_train.select_dtypes(include='object').columns:
    le.fit(x_train[col].unique())
    x_train[col]=le.transform(x_train[col])
    x_test[col]=le.transform(x_test[col])
    with open(f'{col}_label_encoder.pkl', 'wb') as f:
        pickle.dump(le, f)
        
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score ,mean_absolute_error
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

LinearRegression

```python
print("mean_squared_error:",mean_squared_error(y_test,y_pred))
print("r2_score:",r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor(random_state=42,max_depth=8)
dt_model.fit(x_train,y_train)
```

```python
y_pred_dt=dt_model.predict(x_test)
```

DecisionTreeRegressor

```python
print("mean absolute_error:",mean_absolute_error(y_test,y_pred_dt))
print("mean_squared_error:",mean_squared_error(y_test,y_pred_dt))
print("r2_score:",r2_score(y_test,y_pred_dt))
```

```python
from sklearn.preprocessing import MinMaxScaler
my_scaler = MinMaxScaler(feature_range=(0,1))
X_train = my_scaler.fit_transform(x_train)
X_test = my_scaler.transform(x_test)
```

```python
from xgboost import  XGBRegressor
```

```python
model =  XGBRegressor(
        n_estimators=100,       # عدد الأشجار
        learning_rate=0.1,      # تأثير كل شجرة
        max_depth=6,            # أقصى عمق لكل شجرة
        subsample=1.0,          # نسبة العينات المستخدمة لكل شجرة
        colsample_bytree=1.0,   # نسبة الأعمدة المستخدمة لكل شجرة
        gamma=1,                # الحد الأدنى للـloss reduction لتقسيم العقدة
        early_stopping_rounds =10,  # التوقف المبكر بعد 10 جولات بدون تحسن
        random_state=42,
        eval_metric='rmse',  # نوع الخسارة للتصنيف
        verbosity=1
    )
```

```python
model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],   # بيانات الـ validation
    verbose=1
)

```

 XGBRegressor

```python
y_pred_xgbreg = model.predict(x_test)
print(model.score(x_test,y_test))
```

```python
from catboost import  CatBoostRegressor
```

```python
model =  CatBoostRegressor(
        iterations=100,         # عدد الأشجار
        learning_rate=0.1,      # تأثير كل شجرة
        depth=6,                # أقصى عمق لكل شجرة
        l2_leaf_reg=3.0,        # regularization لتقليل overfitting
        loss_function='RMSE',   # نوع الخسارة: RMSE, MAE, Quantile, etc.
        verbose=1,
        random_state=42,
        eval_metric='R2', # مقياس التقييم: RMSE, MAE, R2, etc.
        early_stopping_rounds=10
    )
model.fit(x_train,y_train,eval_set=(x_test, y_test), verbose=1)
```

CatBoostRegressor

```python
y_pred = model.predict(x_test)
print(model.score(x_test,y_test))
```
